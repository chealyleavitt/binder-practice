```python
#  How many different companies are represented in the data set?
len(df['Company Name'].unique())
```

```python
# What is the total number of jobs created for businesses in Queens?
np.sum(df[df['Borough']=='Queens']['Job created'])
```

```python
# How many different unique email domains names are there in the data set?
len(df['company email'].str.split('@').str[-1].unique())
```

```python
# Considering only NTAs with at least 5 listed businesses, 
# what is the average total savings and the total jobs created for each NTA?
sub = df['NTA'].value_counts()
```

```python
ntas = list(sub[sub.values>=5].index)
```

```python
df2 = df[df['NTA'].isin(ntas)]
```

```python
df2 = df2[['NTA','Job created','Total Savings']].groupby('NTA').mean()
```

```python
df2.head()
```

```python
df2.to_csv('NTAsummary.csv')
```

```python
import altair as alt
```

```python
source = df2

alt.Chart(source).mark_point().encode(
    x='Job created',
    y=alt.Y('Total Savings')
)
```

```python
source = df2

alt.Chart(source).mark_point().encode(
    x='Job created',
    y=alt.Y('Total Savings', scale=alt.Scale(type="log"))
)
```

```python
df2['log(Total Savings)'] = [np.log10(x) for x in df2['Total Savings']]
source = df2

alt.Chart(source).mark_bar().encode(
    alt.X("log(Total Savings)", bin=True),
    y='count()',
)
```
